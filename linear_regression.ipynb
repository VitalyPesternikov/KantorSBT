```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
def function_1(x):
    return 2 * x - 1
```

```python
x1 = np.linspace(-1, 1, 500)
y1 = function_1(x1)+np.random.normal(0, 0.1, 500)
plt.figure(figsize=(10, 8))
plt.xlim(-1.1, 1.1)
plt.ylim(-3.5, 1.5)
plt.scatter(x1, y1, s = 4)
plt.show()
```

```python
x2 = np.linspace(-1, 1, 100)
y2 = 1 + np.random.normal(0, 0.1, 100)
plt.figure(figsize=(10, 8))
plt.xlim(-1.1, 1.1)
plt.ylim(-3.5, 1.5)
plt.scatter(x2, y2, s = 4)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x1, y1, color='r', s=4, label = 'first array')
plt.scatter(x2, y2, color='b', s=4, label = 'second array')
plt.plot(x1, function_1(x1), color='g', linewidth=2, label = 'grafic')
plt.legend()
plt.grid()
plt.show()
```

```python
def loss_function_MSE(coeffs):
    k, b = coeffs
    return sum((k * x + b - y)**2)
```

```python
x = np.hstack((x1, x2))
y = np.hstack((y1, y2))
k_1 = minimize(loss_function_MSE, [0, 0]).x[0]
b_1 = minimize(loss_function_MSE, [0, 0]).x[1]
print k_1, b_1
y_MSE = x * k_1 + b_1
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='r', s=4, label = 'all dots')
plt.plot(x, function_1(x), color='g', linewidth = 2, label = 'grafic')
plt.plot(x, y_MSE, color='b',linewidth = 2, label = 'regression MSE')
plt.legend()
plt.grid()
plt.show()
```

```python
def loss_function_MAE(coeffs):
    k, b = coeffs
    return sum(abs(k * x + b - y))
```

```python
k_2 = minimize(loss_function_MAE, [0, 0]).x[0]
b_2 = minimize(loss_function_MAE, [0, 0]).x[1]
print k_2, b_2
y_MAE = x * k_2 + b_2
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='r', s=4, label = 'all dots')
plt.plot(x, function_1(x), color='g', linewidth = 2, label = 'grafic')
plt.plot(x, y_MAE, color='b',linewidth = 2, label = 'regression MAE')
plt.legend()
plt.grid()
plt.show()
```
